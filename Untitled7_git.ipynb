```python
import numpy as np
import tweepy
import re, datetime,json, pandas as pd
consumer_key = ''
consumer_secret = ''
access_token = '-'
access_token_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

Twitter Api kullanılarak Hillary Clinton kişisinin tweet bilgisinin alınması

Twitter'dan çekilen tweet bilgisini otomatik okumak için gerekli şablonu içinde barındıran class yapısıdır.

```python
#TweetMiner function from Nisanur Bulut

class TweetMiner(object):

    
    def __init__(self, api, result_limit = 20):
        
        self.api = api        
        self.result_limit = result_limit
        

    def mine_user_tweets(self, user="HillaryClinton", mine_retweets=False, max_pages=20):

        data           =  []
        last_tweet_id  =  False
        page           =  1
        
        while page <= max_pages:
            
            if last_tweet_id:
                statuses   =   self.api.user_timeline(screen_name=user, count=self.result_limit, max_id=last_tweet_id - 1, include_rts=mine_retweets)                               
                          
            else:
                statuses   =   self.api.user_timeline(screen_name=user, count=self.result_limit, include_rts=mine_retweets)
               
               
            
            for item in statuses:                              
                item1=json.loads(json.dumps(item._json)) 

                # try except alanı
                # retweet alanı sıfır ise go to except 
                try:
                   
                    mined = {
                        'tweet_id':        item1['id'],
                        'handle':          item1['user']['screen_name'],
                        'retweet_count':   item1['retweet_count'],
                        'text':            item1['text'],
                        'mined_at':        datetime.datetime.now(),
                        'created_at':      item1['created_at'],
                    }
                
                except:
                        mined = {
                        'tweet_id':        item1['id'],
                        'handle':          item1['user']['screen_name'],
                        'retweet_count':   0,
                        'text':            item1['text'],
                        'mined_at':        datetime.datetime.now(),
                        'created_at':      item1['created_at'],
                    }
                
                last_tweet_id = item1['id']
                data.append(mined)
                
            page += 1
            
        return data
```

```python
# Veri çekimi
miner = TweetMiner(api, result_limit=2)
hillary = miner.mine_user_tweets(user="HillaryClinton")
donald = miner.mine_user_tweets(user="realDonaldTrump")
```

```python
#cleaning text data using textacy
from textacy.preprocess import preprocess_text

tweet_text = tweets['text'].values
clean_text = [preprocess_text(x, fix_unicode=True, lowercase=True, no_urls=True, no_emails=True, no_phone_numbers=True, no_currency_symbols=True,
                              no_punct=True, no_accents=True)
              for x in tweet_text]
               
```

```python
#hedefin oluşturulmasu
y = tweets['handle'].map(lambda x: 1 if x == 'realDonaldTrump' else 0).values
#max olan olasılık alınır
print(pd.Series(y).value_counts(normalize=True))
print(max(pd.Series(y).value_counts(normalize=True)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

#Vektörleme  with TF-IDF Vectorizer ve x matrix oluşturulması
tfv = TfidfVectorizer(ngram_range=(2,4), max_features=2000)
X = tfv.fit_transform(clean_text).todense()
print (X.shape)#boyut gösterimi
```

```python
from sklearn.model_selection import GridSearchCV

#parametre ayarı yaparak başarının yükselmesi amacıyla GridSearchCV kullanılmıştır

lr = LogisticRegression()
params = {'penalty': ['l1', 'l2'], 'C':np.logspace(-5,0,100)}
#Grid searching to find optimal parameters for Logistic Regression
gs = GridSearchCV(lr, param_grid=params, cv=10, verbose=1)
gs.fit(X, y)
```

```python
print (gs.best_params_)
print (gs.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(LogisticRegression(), X, y, cv=10)

print (accuracies.mean())
print (1-y.mean())
```
